```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_csv('car_sale.csv')
df
```

```python
original=df.copy()

#this list will contain names of attributes to drop
remove_list=[]
```

```python
# Data cleaning and exploration 
```

```python
#removing na values
df.info()
df=df.dropna()
```

```python
#we remove model attribute as it contains too many distinct attributes which dont show a lot of significance
df['Model'].value_counts(normalize=True)
remove_list.append('Model')
```

```python
#we see that the cc_value of engine is
df['Engine']
df['Engine']=df['Engine'].apply(lambda x:float(x.split(' ')[0]))
```

```python
#we have removed the cc from the Engine Attribute and made it numeric
df=df.rename(columns={'Engine':'Engine(cc)'})
df
```

```python
#similarly we now have to deal with categories like Max Power
df['Max Power']=df['Max Power'].str.replace('bhp',"")
df['Max Power']=df['Max Power'].str.replace('rpm',"")
df['Max Power']=df['Max Power'].str.replace(' ',"")
df['Max Power']=df['Max Power'].str.replace('@'," ")
temp=df['Max Power'].str.split(" ")
```

```python
#with this we drop max power attribute and two other attributes called Max Power(BHP) and Max Power(RPM)
temp1=temp.apply(lambda x:float(x[0]))
temp2=temp.apply(lambda x:float(x[1]))

df['Max Power(BHP)']=temp1
df['Max Power(RPM)']=temp2
df=df.drop(['Max Power'],axis=1)
df
```

```python
#similar to Max Power we now repeat this with the attribute Max Torque
df['Max Torque']=df['Max Torque'].str.replace('Nm',"")
df['Max Torque']=df['Max Torque'].str.replace('rpm',"")
df['Max Torque']=df['Max Torque'].str.replace(' ',"")
df['Max Torque']=df['Max Torque'].str.replace('@'," ")
temp=df['Max Torque'].str.split(" ")
```

```python
#with this we drop max Torque attribute and two other attributes called Max Torque(NM) and Max Torque(RPM)
temp1=temp.apply(lambda x:float(x[0]))
temp2=temp.apply(lambda x:float(x[1]))
df['Max Torque(NM)']=temp1
df['Max Torque(RPM)']=temp2
df=df.drop(['Max Torque'],axis=1)
df
```

```python
df.info()
```

```python
# we apply Ordinal Encoding to the 'Owner' attribute as there is a chance it 
#if the car is 'sold a second time it might have more value than a car being sold for the third time'
df['Owner'].value_counts(normalize=True)
```

```python
#ordinal encoidng using lambda and apply
df['Owner']=df['Owner'].apply(lambda x: 1 if x=='First' else x)
df['Owner']=df['Owner'].apply(lambda x: 2 if x=='Second' else x)
df['Owner']=df['Owner'].apply(lambda x: 3 if x=='Third' else x)
df['Owner']=df['Owner'].apply(lambda x: 4 if x=='UnRegistered Car' else x)
```

```python
#as we can see we have transformed the data
df['Owner'].value_counts()
```

```python
df['Fuel Type'].value_counts(normalize=True)
```

```python
#cng+cng might mean it runs only on cng so we dont need another classification
df['Fuel Type']=df['Fuel Type'].apply(lambda x: "CNG" if (x=="CNG + CNG") else x)
```

```python
df['Transmission'].value_counts(normalize=True)
```

```python
df['Location'].value_counts(normalize=True)
```

```python
df['Drivetrain'].value_counts(normalize=True)
```

```python
df['Color'].value_counts(normalize=True)
```

```python
df=df.drop(['Model'],axis=1)
```

```python
df_copy=df.copy()
```

```python
#we apply one hot encoding to the rest
df=pd.get_dummies(df)
```

```python
#data cleaing complete
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

X=df.drop(['Price'],axis=1)
y=df['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.30, random_state=42)

lr=LinearRegression()
std_scl=StandardScaler()
```

```python
#training a basic Linear Regression Model with Standard Scaler

simple_reg=Pipeline([('std_scl',std_scl),('lr',lr)])
simple_reg.fit(X_train,y_train)
print("R2 score of Simple Linear Regression\nwith Standard Scaler :",simple_reg.score(X_test,y_test))
y_pred=simple_reg.predict(X_test)
sns.distplot(y_test)
sns.distplot(y_pred)
```

```python
#training a basic Linear Regression Model without Standard Scaler

lr.fit(X_train,y_train)
print("R2 score of Simple Linear Regression\nwithout Standard Scaler :",lr.score(X_test,y_test))
y_pred=lr.predict(X_test)
sns.distplot(y_test)
sns.distplot(y_pred)

#we can see that we have better results for Linear Regression without standard Scaler So we continue without it 
```

```python
#we try the Linear regression with log and polynomial functions to see if it affects the performance

#polynomial with degree 3

from sklearn.preprocessing import PolynomialFeatures

ply=PolynomialFeatures(degree=3)
simple_reg=Pipeline([('ply',ply),('lr',lr)])
simple_reg.fit(X_train,y_train)
print("R2 score of Simple Linear Regression\nwith Polynomial degree 3  :",simple_reg.score(X_test,y_test))
y_pred=simple_reg.predict(X_test)
sns.distplot(y_test)
sns.distplot(y_pred)
```

```python
#polynomial with degree 2

ply=PolynomialFeatures(degree=2)
simple_reg=Pipeline([('ply',ply),('lr',lr)])
simple_reg.fit(X_train,y_train)
print("R2 score of Simple Linear Regression\nwith Polynomial degree 2 :",simple_reg.score(X_test,y_test))
y_pred=simple_reg.predict(X_test)
sns.distplot(y_test)
sns.distplot(y_pred)
```

```python
#linear regression with log

simple_reg=Pipeline([('lr',lr)])
simple_reg.fit(X_train,np.log(y_train))
print("R2 score of Simple Linear Regression\nwith Log :",simple_reg.score(X_test,y_test))
y_pred=simple_reg.predict(X_test)
sns.distplot(y_test)
sns.distplot(y_pred)

#we see that model performs worse
```

```python
#so far plain/ simple linear regression has given us the best model, so we look at its parameters(Co-effs)
lr.fit(X_train,y_train)
model_coef=lr.coef_
coef_df=pd.DataFrame(zip(X.columns, model_coef))
coef_df
```

```python
#Ridge Regression 
from sklearn.linear_model import Ridge
rid=Ridge(alpha=200)
rid.fit(X_train,y_train)
rid.score(X_test,y_test)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
las=Lasso(alpha=19000)
las.fit(X_train,y_train)
las.score(X_test,y_test)
```

```python
#Elastic Net Regression
from sklearn.linear_model import ElasticNet
las=ElasticNet(alpha=19000,l1_ratio=0.5)
las.fit(X_train,y_train)
las.score(X_test,y_test)
```
